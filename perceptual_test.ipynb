```python
import numpy as np
import matplotlib.image as mpimg
from matplotlib import pyplot as plt
```

```python
test_1 = [1,3]
test_2 = [6,4]
test_3 = [15,7]
test_4 = [5,9]
test_5 = [16,20]
test_6 = [2,18]
test_7 = [14,17]
test_8 = [19,13]
test_9 = [8,11]
test_10 = [12,10]
```

```python
def show(test):    
    sample_1 = './amt/sampl' + str(test[0]) + '.png'
    sample_2 = './amt/sampl' + str(test[1]) + '.png'
    smpl1 = mpimg.imread(sample_1)
    smpl2 = mpimg.imread(sample_2)
    
    fig = plt.figure(figsize=(8, 8), facecolor='w')
    plt.subplot(2, 2, 1)
    plt.imshow(smpl1)

    plt.axis('off')
    plt.subplot(2, 2, 2)
    plt.imshow(smpl2)

    plt.axis('off')
    plt.show()
```

```python
show(test_10)
```
